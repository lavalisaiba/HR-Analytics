```python
#Import Necessary Libraries
import pandas as pd
import seaborn as sns
```

```python
#Loading the Dataset
data = pd.read_csv("HR-Employee-Attrition.csv")
```

**Data Anomilies**

```python
data.columns
```

Age: The age of the employees in the dataset.

Attrition: This column has been modified to contain binary values, with 1 indicating "Yes" (employee has left the company) and 0 indicating "No" (employee has not left).

BusinessTravel: The frequency of business travel, categorized as "Travel_Rarely," "Travel_Frequently," or "Non-Travel."

DailyRate: The daily rate of pay for employees.

Department: The department in which the employee works, such as Sales, Research & Development, or Human Resources.

DistanceFromHome: The distance of the employee's home from the workplace.

Education: The level of education of the employee, ranging from 1 (Below College) to 5 (Doctorate).

EducationField: The field of education or specialization of the employee, such as Life Sciences, Medical, or Technical Degree.

EnvironmentSatisfaction: The satisfaction level of employees with their working environment, ranging from 1 (Low) to 4 (High).

Gender: The gender of the employee, categorized as Male or Female.

HourlyRate: The hourly rate of pay for employees.

JobInvolvement: The level of job involvement or engagement of employees, ranging from 1 (Low) to 4 (High).

JobLevel: The level of the employee's job within the company hierarchy.

JobRole: The specific role or position of the employee within the company, such as Sales Executive, Research Scientist, or Laboratory Technician.

JobSatisfaction: The satisfaction level of employees with their job, ranging from 1 (Low) to 4 (High).

MaritalStatus: The marital status of employees, categorized as Single, Married, or Divorced.

MonthlyIncome: The monthly income of employees.

MonthlyRate: The monthly rate of pay for employees.

NumCompaniesWorked: The number of companies the employee has worked for prior to the current job.

OverTime: Indicates whether the employee works overtime, categorized as Yes or No.

PercentSalaryHike: The percentage of salary increase for employees.

PerformanceRating: The performance rating of employees, typically 3 or 4.

RelationshipSatisfaction: The satisfaction level of employees with their relationships at work, ranging from 1 (Low) to 4 (High).

StockOptionLevel: The level of stock options or benefits offered to employees.

TotalWorkingYears: The total number of years the employee has worked in their career.

TrainingTimesLastYear: The number of times the employee received training in the last year.

WorkLifeBalance: The balance between work and personal life, ranging from 1 (Bad) to 4 (Good).

YearsAtCompany: The total number of years the employee has been with the company.

YearsInCurrentRole: The number of years the employee has been in their current role.

YearsSinceLastPromotion: The number of years since the employee's last promotion.

YearsWithCurrManager: The number of years the employee has been with their current manager

**Correlation Matrix**

```python
import matplotlib.pyplot as plt
# Load the dataset
data = pd.read_csv("HR-Employee-Attrition.csv")

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(15, 12))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

**Overall Attrition Rate**

```python
# Calculate the overall attrition rate
total_employees = len(data)
attrition_count = data["Attrition"].sum()  # Count of employees who left (Attrition = 1)

```

```python
attrition_count
```

```python
attrition_rate = attrition_count / total_employees
print("Overall Attrition Rate: {:.2%}".format(attrition_rate))
```

# Overall Attrition Rate: 16.12%

**Attrtion By Department**

```python
department_attrition = data.groupby("Department")["Attrition"].sum().reset_index()
print(department_attrition)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(department_attrition["Department"], department_attrition["Attrition"], color='skyblue')
plt.title('Total Attritions by Department')
plt.xlabel('Department')
plt.ylabel('Total Attritions')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

Highest Attrtion of 133 in Research & Development Department

**Attrition By Marital Status and Gender**

```python
marital_gender_attrition=data.groupby(["MaritalStatus","Gender"])["Attrition"].sum().reset_index()
print(marital_gender_attrition)
```

```python
pivot_marital_gender_attrition = data.pivot_table(index=["MaritalStatus", "Gender"],
                                                 values="Attrition", aggfunc="sum").reset_index()
plt.figure(figsize=(10, 6))
plt.bar(pivot_marital_gender_attrition["MaritalStatus"] + " - " + pivot_marital_gender_attrition["Gender"],
        pivot_marital_gender_attrition["Attrition"], color='skyblue')
plt.title('Total Attritions by MaritalStatus and Gender')
plt.xlabel('MaritalStatus - Gender')
plt.ylabel('Total Attritions')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

Attrtion count is Highest in Single Male abouyt 73, and Married Man about 53

Where as Attrition count in Single Femal is 47 and in married about 31.

**Average Age of Employees who are departing rom company**

```python
departing_employees = data[data["Attrition"] == 1]
# Calculate the average age of departing employees
average_age_departing = departing_employees["Age"].mean()
print("Average Age of Departing Employees: {:.2f}".format(average_age_departing))
```

**Attrition by Monthly Income**

```python
attrition_yes = data[data["Attrition"] == 1]
attrition_no = data[data["Attrition"] == 0]
# Create a histogram to compare the distribution of monthly incomes
plt.figure(figsize=(10, 6))
plt.hist(attrition_yes["MonthlyIncome"], bins=20, alpha=0.5, color='red', label='Attrition: Yes')
plt.hist(attrition_no["MonthlyIncome"], bins=20, alpha=0.5, color='blue', label='Attrition: No')
plt.title('Attrition by Monthly Income')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# CONCLUSION

The dataset reveals a significant attrition rate of 16.21%.

 Attrition is higher in the Research and Development department, among single males, and employees who work overtime. Average age and tenure of departing employees are 33 years and 8 years, respectively. Bachelor's degree holders and those with moderate job satisfaction show higher attrition. Further analysis is needed to understand and address the underlying reasons for attrition and develop retention strategies.
